# Pandas

```python
import numpy as np
import pandas as pd
```

```python
# Crear matrices con la ayuda de numpy
mat_1 = np.array([1,2,3,4,5])
print(f"El objeto mat_1 es \n{mat_1}")
print(f"\nEl objeto mat_1 es del tipo {type(mat_1)}")
```

```python
# Crear matrices con la ayuda de numpy
mat_2 = np.arange(10)
print(f"El objeto mat_2 es \n{mat_2}")
print(f"\nEl objeto mat_2 es del tipo {type(mat_2)}")
```

```python
# Crear matrices con números aleatorios
mat_3 = np.random.uniform(5,6,10)
print(f"El objeto mat_3 es \n{mat_3}")
print(f"\nEl objeto mat_3 es del tipo {type(mat_3)}")
```

```python
mat_4 = np.random.uniform(1,2,10)
print(f"El objeto mat_4 es \n{mat_4}")
print(f"\nEl objeto mat_4 es del tipo {type(mat_4)}")
```

```python
mat_5 = np.random.uniform(-1,1,10)
print(f"El objeto mat_5 es \n{mat_5}")
print(f"\nEl objeto mat_5 es del tipo {type(mat_5)}")
```

```python
# Sumar las matrices 1D de numpy
mat_sum = mat_3 + mat_4 + mat_5
print(f"La suma de las matrices es \n{mat_sum}")
```

```python
# Obteniendo algunas estadísticas de las matrices en numpy
print(f"El mínimo del resultado de la suma de matrices es {mat_sum.min()}")
print(f"El mínimo del resultado de la suma de matrices es {mat_sum.max()}")
print(f"El mínimo del resultado de la suma de matrices es {mat_sum.sum()}")
```

## Series

```python
# Crear series a partir de matrices de 1D de numpy
ser_mat_3 = pd.Series(mat_3)
ser_mat_4 = pd.Series(mat_4)
ser_mat_5 = pd.Series(mat_5)
print(f"El objeto ser_mat_3 es \n{ser_mat_3}")
print(f"\nEl objeto ser_mat_3 es del tipo {type(ser_mat_3)}\n")
print(f"El objeto ser_mat_4 es \n{ser_mat_4}")
print(f"\nEl objeto ser_mat_4 es del tipo {type(ser_mat_4)}\n")
print(f"El objeto ser_mat_5 es \n{ser_mat_5}")
print(f"\nEl objeto ser_mat_5 es del tipo {type(ser_mat_5)}")
```

```python
ser_mat_sum = ser_mat_3 + ser_mat_4 + ser_mat_5
print(f"La suma de las series es \n{ser_mat_sum}")
```

```python
# Describir el objeto ser_mat_sum (Series de Pandas)
ser_mat_sum.describe()
```

```python
# Crear un diccionario (Clave:Valor)
# Claves como cadenas de caracteres
# Valores como Series de Pandas

dc_1 = {"Serie_1":ser_mat_3,
        "Serie_2":ser_mat_4,
        "Serie_3":ser_mat_5,
        "Serie_4":ser_mat_sum
        }

print(f"El objeto dc_1 es \n{dc_1}")
print(f"El objeto dc_1 es del tipo \n{type(dc_1)}")
```

```python
# Crear un DataFrame a partir de un diccionario
data_dc = pd.DataFrame(dc_1)
```

```python
data_dc.head(5)
```

```python
# Obtener las estadísticas de los campos o variables del DataFrame
data_dc.describe()
```

```python
res_data_dc = data_dc.describe().transpose().reset_index()
res_data_dc
```

```python
# Llamar a la variable "std" del DataFrame
res_data_dc["std"]
```

```python
# Calcular una variable nueva en el DataFrame a partir de variables existentes
# Coeficiente de variación
res_data_dc["cv"] = np.round((res_data_dc["std"] / np.abs(res_data_dc["mean"]))*100,2)
res_data_dc
```
